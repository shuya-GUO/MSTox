```python
from rdkit import Chem
import pandas as pd

df = pd.read_csv("output.csv")
def valid_smiles(s):
    try:
        return Chem.MolFromSmiles(s) is not None
    except:
        return False

df = df[df["SMILES"].notna() & (df["SMILES"] != "N/A")]
df = df[df["SMILES"].apply(valid_smiles)]
df.drop_duplicates(subset=["SPECTRUMID"], inplace=True)
df.to_csv("cleaned_metadata.csv", index=False)
print(len(df))
```
